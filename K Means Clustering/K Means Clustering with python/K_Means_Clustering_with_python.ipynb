Data Gathering (Unsupervised)

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4,cluster_std=1.8,
                random_state=101)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,0], c=data[1],cmap="rainbow")
```

Model

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4) #You can add 2,3,4,5,6,7,8 for better understand
model.fit(data[0])
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title("K Means")
ax1.scatter(data[0][:,0], data[0][:,1], c=model.labels_, cmap="rainbow")

ax2.set_title("Orginal")
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap="rainbow")
```
